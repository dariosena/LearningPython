```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pickle
```

```python
print(scipy.__version__)
print(scipy.__file__)
```

```python
def loadlena():
    f = open('lena.dat','rb')
    lena = np.array(pickle.load(f))
    f.close()
    return lena
```

```python
# Set the resize factors
yfactor = 2
xfactor = 3

#Lena's dimensions
LENA_X = 512
LENA_Y = 512
```

```python
lena = loadlena()
```

```python
np.testing.assert_equal((LENA_X, LENA_Y), lena.shape)
```

```python
resized = lena.repeat(yfactor, axis=0).repeat(xfactor, axis=1)
```

```python
plt.subplot(211)
plt.title("Lena")
plt.axis("off")
plt.imshow(lena)

plt.subplot(212)
plt.title("Resized")
plt.axis("off")
plt.imshow(resized)
plt.show()
```
